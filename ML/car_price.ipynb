```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/ML by study mart/4.car_price/carprices.csv")
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(["Car Model"],axis="columns")
final
```

```python
y = df['Sell Price']
y
```

```python
x = final.drop(['Sell Price'],axis='columns')
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[69000,6,0,1,0,0]])
```
